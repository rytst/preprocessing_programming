# 9-1 Vertical to horizontal

## Pandas

```python
import pandas as pd
```

```python
pd_reservation = pd.read_parquet("../../data/reservation.parquet")
```

```python
pd_reservation.head()
```

```python
pd.pivot_table(pd_reservation, index="hotel_id", columns="people_num", aggfunc="size")
```

```python
(
    pd_reservation
    .groupby(["hotel_id", "people_num"]).size()
    .unstack()
)
```

## Polars

```python
import polars as pl
```

```python
pl_reservation = pl.read_parquet("../../data/reservation.parquet")
```

```python
pl_reservation.head()
```

```python
pl_reservation.pivot(
    values="reservation_id",
    index="hotel_id",
    on="people_num",
    aggregate_function="len",
    sort_columns=True
).sort("hotel_id")
```
