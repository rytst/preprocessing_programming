# 7-1 One-to-one and many-to-one

## Extract business hotel and people number = 1 reservation data

### Pandas

```python
import pandas as pd
```

```python
pd_reservation = pd.read_parquet("../../data/reservation.parquet")
pd_hotel       = pd.read_parquet("../../data/hotel.parquet")
```

```python
pd_reservation.head()
```

```python
pd_hotel.head()
```

```python
(
    pd_reservation
    .query("people_num == 1")
    .merge(
        pd_hotel.query("hotel_type == 'ビジネスホテル'")["hotel_id"],
        on="hotel_id",
        how="inner",
    )
)
```

### Polars

```python
import polars as pl
```

```python
pl_reservation = pl.read_parquet("../../data/reservation.parquet")
pl_hotel       = pl.read_parquet("../../data/hotel.parquet")
```

```python
pl_reservation.head()
```

```python
pl_hotel.head()
```

```python
(
    pl_reservation
    .filter(pl.col("people_num") == 1)
    .join(
        pl_hotel.filter(pl.col("hotel_type") == "ビジネスホテル").select("hotel_id"),
        on="hotel_id",
        how="inner"
    )
)
```

```python
# semi
(
    pl_reservation
    .filter(pl.col("people_num") == 1)
    .join(
        pl_hotel.filter(pl.col("hotel_type") == "ビジネスホテル"),
        on="hotel_id",
        how="semi",
    )
)
```
