# 5-2 Extract columns by condition

## Column name starts with "tag_"

### Pandas

```python
import pandas as pd
```

```python
pd_hotel = pd.read_parquet("../../data/hotel.parquet")
```

```python
pd_hotel.head()
```

```python
# Partial matching
pd_hotel.loc[:, lambda df: df.columns.str.startswith("tag_")]
```

### Polars

```python
import polars as pl
```

```python
pl_hotel = pl.read_parquet("../../data/hotel.parquet")
```

```python
pl_hotel.head()
```

```python
pl_hotel.select(pl.col("^tag_.*$"))
```

## Null Value

### Pandas

```python
pd_customer = pd.read_parquet("../../data/customer.parquet")
```

```python
pd_customer.head()
```

```python
pd_customer.loc[:, lambda df: df.isnull().any()]
```

### Polars

```python
pl_customer = pl.read_parquet("../../data/customer.parquet")
```

```python
pl_customer.head()
```

```python
cols = [col.name for col in pl_customer.select(pl.all().is_null()) if col.any()]
cols
```

```python
pl_customer.select(pl.col(cols))
```

## Numeric value

### Pandas

```python
pd_reservation = pd.read_parquet("../../data/reservation.parquet")
```

```python
# select all columns which have numerical value 
pd_reservation.select_dtypes("number") # `select_dtypes` function
```

```python
# select all columns which have object type
pd_reservation.select_dtypes("object")
```

```python
# select all columns which have datetime type
pd_reservation.select_dtypes(["datetime", "datetimetz"])
```

### Polars

```python
pl_reservation = pl.read_parquet("../../data/reservation.parquet")
```

```python
pl_reservation.select(pl.col(pl.Int64))
```
