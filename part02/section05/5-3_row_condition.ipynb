# 5-3 Extract rows by condition

## Extract 2-4 people-num reservation data

### Pandas 

```python
import pandas as pd
```

```python
pd_reservation = pd.read_parquet("../../data/reservation.parquet")
```

```python
pd_reservation.head()
```

```python
# query function
pd_reservation.query("2 <= people_num <= 4")
```

```python
pd_reservation.loc[lambda df: df.people_num.between(2,4)]
```

### Polars

```python
import polars as pl
```

```python
pl_reservation = pl.read_parquet("../../data/reservation.parquet")
```

```python
pl_reservation.head()
```

```python
pl_reservation.filter(pl.col("people_num").is_between(2, 4))
```
